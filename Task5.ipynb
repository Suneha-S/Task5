```python
#Pair plot
df = pd.read_csv(r'C:\Users\l\OneDrive\Desktop\train.csv')
# Encoding categorical variables for plotting
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
# Selecting relevant features for pair plot
selected_features = ['Sex', 'Age', 'Fare', 'Survived']
sns.pairplot(df[selected_features], hue='Survived', palette='Set2')
# Displaying the plot
plt.suptitle("Pair Plot of Titanic Dataset", y=1.02)
plt.show()
```

```python
#Histogram
df[['Age', 'Fare', 'SibSp', 'Parch']].hist(bins=20, figsize=(10, 8), color='skyblue')
plt.suptitle('Histograms')
plt.tight_layout()
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='Survived', y='Age', data=df, palette='Set2',legend=False)
plt.title('Age vs Survival')
plt.subplot(1, 2, 2)
sns.boxplot(x='Survived', y='Fare', data=df, palette='Set2',legend=False)
plt.title('Fare vs Survival')
plt.tight_layout()
plt.show()
```
